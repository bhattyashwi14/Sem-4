```python
dff["Total"]=dff.apply(sum,axis=1)
dff
```

```python
def grade(x):
    if x>300:
        return "A+"
    elif x>250:
        return "A"
    elif x>200:
        return "B+"
    elif x>150:
        return "B"
    else:
        return "F"
    
```

```python
dff["Grade"]=dff["Total"].apply(grade)
dff
```

```python
def result(x):
    if x["A"]>=35 and x["B"]>=35 and x["C"]>=35 and x["D"]>=35 and x["E"]>=35:
        return "Pass"
    else:
        return "Fail"
```

```python
dff["Result"]=dff.apply(result,axis=1)
dff
```

```python
dff["Result"]=dff.apply(lambda x: "Pass" if (x["A"]>=35 and x["B"]>=35 and x["C"]>=35 and x["D"]>=35 and x["E"]>=35)
                        else "Fail"  ,axis=1)
dff
```

```python
def bonus(x):
    return x+2
```
