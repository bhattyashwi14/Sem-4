```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id")
# takes only the common value of the entered column
# here df1 has id 6 and df2 has id 5, so those two enteries won't be present while merging
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
pd.merge(df1,df2,on="id",how="left")
# all the data of first data frame and common entries of second
```

```python
pd.merge(df1,df2,on="id",how="right")
# all the data of second data frame and common entries of first
```

```python
pd.merge(df1,df2,how="cross")
```

```python
# pd.merge(df1,df2,on="id",how="corss")
# it will get confused and give error when we specify column
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=("_df1","_df2"))
```

```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"in":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,8]})
```

```python
df1.merge(df2,left_on="in",right_on="id")
# only common values will appear
```

```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"in":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,8],"fac":["vha","nas","vby","mss","dps"]})
```

```python
df1.merge(df2,left_on="name",right_on="fac")
```

```python
df1=pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"id":[3,2,1,4,6]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,8],"fac":["vha","nas","vby","mss","dps"]})
```

```python
df1.merge(df2,left_on="name",right_on="fac",how="outer",suffixes=("_b1_sem3","_b1_sem4"))
```
