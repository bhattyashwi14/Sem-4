```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame([[1,np.nan,3,4],[2,4,np.nan,8],[np.nan,4,5,9],[np.nan,1,np.nan,3],[np.nan,np.nan,np.nan,np.nan],[1,1,1,1]],
                columns=["A","B","C","D"])
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
# even a single value in row is null, it'll drop the entire row
```

```python
df.dropna(how="any",axis=1)
# drops each column having even single nan value
```

```python
df.dropna(how="all")
# removes only the row which has all values NaN
```

```python
df.dropna(subset=["A"])
# removes the rows in which A column is null
```

```python
df.dropna(how="all",subset=["B","D"])
```

```python
df.dropna(thresh=3)
# if filling value (non NaN values) is equal to 3 or greater, then that row won't be dropped
```

```python
df.dropna(thresh=2)
# if filling value (non NaN values) is equal to 2 or greater, then that row won't be dropped
```

```python
df.fillna(df.mean())
# mean is column wise
```

```python
df["A"]=df["A"].fillna(df["A"].mean())
df
```
