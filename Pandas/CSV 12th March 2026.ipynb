```python
Q1=dff.mpg.quantile(0.25)
Q1
```

```python
Q3=dff.mpg.quantile(0.75)
Q3
```

```python
dff.mpg.describe()
```

```python
IQR=Q3-Q1
IQR
```

```python
HL=Q3+1.5*IQR
LL=Q1-1.5*IQR
print("HL:",HL)
print("LL:",LL)
```

```python
wo=dff[(dff["mpg"]<HL) & (dff["mpg"]>LL)]
wo
# without outliers data
```

```python
sns.boxenplot(wo.mpg)
```

```python
sns.boxenplot(wo["horsepower"])
```
