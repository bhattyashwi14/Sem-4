```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default="notebook"
```

```python
df=pd.DataFrame({"dep":["hr","it","hr","it","hr"],"age":[20,25,30,22,10],"sal":[100,500,200,300,100],
                 "sex":["F","M","M","F","F"]})
df
```

```python
px.histogram(data_frame=df,x="age")
```

```python
px.histogram(data_frame=df,x="age", color="sex")
```

```python
px.histogram(data_frame=df,x="age", color="sex", nbins=30)
```

```python
px.histogram(data_frame=df,x="age", color="sex",pattern_shape="dep")
```

```python
px.histogram(data_frame=df,x="dep", y="sal")
```

```python
px.histogram(data_frame=df,x="dep", y="sal",histfunc="max")
```

```python
px.histogram(data_frame=df,x="dep", y="sal",histfunc="avg")
```

```python
px.histogram(data_frame=df,x="dep", y="sal",histfunc="count",color="sex")
```

```python
import networkx as nx
```

```python
G=nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,1)
nx.draw(G,with_labels=True)
# gives a new pattern everytime we run it
```

```python
nx.draw(G,with_labels=True,node_size=1000,node_color="cyan")
```

```python
G=nx.DiGraph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,1)
nx.draw(G,with_labels=True,node_size=1000)
```

```python
G=nx.DiGraph()
G.add_nodes_from([1,2,3])
G.add_edges_from([(1,2),(2,3),(3,1),(1,1)])
nx.draw(G,with_labels=True,node_size=1000)
```
