# Q10. Using apply(), calculate the defect rate for each row:
defect_rate=defective_units/units_produced


```python
def rate(x):
    defect_rate=x["defective_units"]/x["units_produced"]
    return defect_rate
df.apply(rate,axis=1)
```

# Q11. Using groupby().apply(), find the profit range (max − min) for each factory.


```python
def profit_range(x):
    l=x["profit"].max()-x["profit"].min()
    return l

df.groupby("factory").apply(profit_range)
```

# Q12. Using groupby().apply(), find the row with maximum units_produced in each product_category.


```python
def max_units(x):
    return x["units_produced"].max()
df.groupby("product_category").apply(max_units)
```

# Q13. Use apply() to create a new column efficiency_score:
efficiency=units_produced/production_time_hr


```python
def efficiency(x):
    eff=x["units_produced"]/x["production_time_hr"]
    return eff
df["efficiency"]=df.apply(efficiency,axis=1)
df
```
